```python
%matplotlib inline
import os, glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.optimize import curve_fit
import scipy.optimize as opt
import matplotlib.lines as mlines
import pandas as pd
import csv
from skimage import io
import matplotlib
from matplotlib.backends.backend_pdf import PdfPages
matplotlib.rcParams.update({'font.size': 32, 'font.family':'sans-serif', 'font.sans-serif':'Arial', 'pdf.fonttype':42})

corr_range = np.array([984,982,985,983,985,983,984])
uncorr_range = np.array([971,943,937,903,889,845,818])
power = np.array([50,100,150,200,250,300,350])

fig, ax = plt.subplots(figsize = (12,10) )


black_set = ax.plot(power,uncorr_range/np.max(corr_range)*100,'.-', color = (0,0,0),markersize=25,linewidth=2.5,label='factory')
red_set = ax.plot(power,corr_range/np.max(corr_range)*100, '.-',color = (1,0,0),markersize=25,linewidth=2.5,label='optimized')

ax.set_xlim(40,360)
ax.set_ylim(0,105)

ax.set_ylabel('relative scan range (%)',fontsize=34)
ax.set_xlabel('average laser power (mW)',fontsize=34)

#################################################################

ax.legend(fontsize = 24,loc=3)
#################################################################
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
with PdfPages('K:/Farinella_2024_repo/processed_data/Fig2/Fig2e.pdf') as pdf:
    pdf.savefig(fig,bbox_inches='tight',pad_inches=0.05,dpi=600)  # saves the current figure into a pdf page
```
